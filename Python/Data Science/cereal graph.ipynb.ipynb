```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cereal = pd.read_csv('cereal.csv')
```

```python
cereal.head()
```

```python
cereal.set_index('name', inplace=True)
```

```python
cereal.head()
```

```python
calories = cereal.calories
sodium = cereal.sodium
```

```python
#color bar will be the cups
cups = []
for x in cereal.cups:
    cups.append(x)
len(cups)
```

```python
cups
```

```python
plt.scatter(calories, sodium, c = cups,
           cmap = 'plasma')
plt.title("Calories & sodium")
plt.xlabel("Calories")
plt.ylabel("Sodium")
plt.legend("f")
cbar = plt.colorbar()
plt.show()
```

The color map represents the cups per serving size
